```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../datasets/police_deaths_USA_v6.csv")
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
df = df[df['Age'] != 1661]
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
# Eliminar filas con valores nulos en columnas importantes
df_cleaned = df.dropna(subset=['Age', 'End_Of_Watch', 'Cause'])
print(df.isnull().sum())
```

```python
#quitar duplicados
df_cleaned = df_cleaned.drop_duplicates()
print(df.isnull().sum())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
#columnas numéricas
df.hist(bins=30, figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Cause', data=df)
plt.xticks(rotation=45)
plt.title('Count of Causes of Death')
plt.show()
```

```python
df['Tour'].unique()
```

```python
df['Tour'] = df['Tour'].fillna("0 tiempo")
df['Tour'].unique()
```

```python
# Seleccionar solo las columnas numéricas correctamente
numeric_columns = df.select_dtypes(include=['float64', 'int64'])

# Matriz de correlación
corr_matrix = numeric_columns.corr()

# Mapa de calor de la matriz de correlación
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()

#sin mucho sentido solo hay 1
```

```python
# Contar valores para cada categoría en una columna específica
for column in df.select_dtypes(include=['object']).columns:
    print(df[column].value_counts())
    print("\n")
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Day_Of_Week', y='Age', data=df)
plt.xticks(rotation=45)
plt.show()
```
